```python
analyzer = PychometricsAnalyzer(
    api_key= "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",
    model_name="google/gemma-3n-e2b-it:free",
)

results, metadata, errors = analyzer.analyze_directory(
    input_dir=sample_dir,
    codebook_path=codebook_path,
    output_dir="my_analysis",
)
```

```python
import pandas as pd

rows = []
for doc_id, doc in results.items():
    for instance in doc.get("instances", []):
        rows.append({
            "construct": instance.get("construct"),
            "speaker_id": instance.get("speaker_id"),
            "quote": instance.get("quote"),
            "quote_index": instance.get("quote_index"),
            "confidence": instance.get("confidence"),
            "document_id": doc_id
        })

df = pd.DataFrame(rows)
```
